```python
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("MinTemperatures")
sc = SparkContext(conf = conf)
```

```python
def parseLine(line):
    fields = line.split(',')
    stationID = fields[0]
    entryType = fields[2]
    temperature = float(fields[3]) * 0.1 * (9.0 / 5.0) + 32.0
    return (stationID, entryType, temperature)
```

```python
lines = sc.textFile("../data/1800.csv")
```

```python
parseLines = lines.map(parseLine)
```

```python
minTemps = parseLines.filter(lambda x: "TMIN" in x[1])
```

```python
stationTemps = minTemps.map(lambda x: (x[0], x[2]))
```

```python
minTemps = stationTemps.reduceByKey(lambda x, y : min(x, y))
```

```python
results = minTemps.collect()
```

```python
for result in results:
    print(result[0] + "\t{:.2f}F".format(result[1]))
```

```python
maxTemp = parseLines.filter(lambda x: "TMAX" in x[1])
```

```python
maxTemp.collect()[:5]
```

```python
stationTemps = maxTemp.map(lambda x: (x[0], x[2]))
```

```python
stationTemps.collect()[:5]
```

```python
maxTemp = stationTemps.reduceByKey(lambda x, y: max(x, y))
```

```python
results = maxTemp.collect()

for result in results:
    print(result[0] + "\t{:.2f}F".format(result[1]))
```
